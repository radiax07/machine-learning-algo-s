```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("dataset.csv", delimiter=",", skiprows=1)
X = data[: , :3]
y = data[:, 3]

y = y.reshape(-1,1)
```

```python
X = (X - np.mean(X, axis=0)) / np.std(X, axis=0) # the datapoints are now in a particular range
```

```python
plt.figure(figsize=(8,6))

plt.scatter(X[:, 0], y)

plt.xlabel("Feature (hours_studied)")
plt.ylabel("Target (passed)")
plt.title("Hours Studied vs Y")

plt.show()
```

```python
plt.figure(figsize=(8,6))

plt.scatter(X[:, 1], y)

plt.xlabel("Feature (attendance)")
plt.ylabel("Target (passed)")
plt.title("Attendance vs Y")

plt.show()
```

```python
plt.figure(figsize=(8,6))

plt.scatter(X[:, 2], y)

plt.xlabel("Feature (sleep_hours)")
plt.ylabel("Target (passed)")
plt.title("Sleep Hours vs Y")

plt.show()
```

```python
def hypothesis(w, X, b): 
    z = np.dot(X,w) + b
    return 1/(1 + np.exp(-z))
```

```python
def loss(y_pred, y, X):
    n = len(y)
    J = -1/n * np.sum(y*np.log(y_pred) + (1-y) * np.log(1-y_pred))

    return J
```

```python
def gradient(y_pred, y, X):
    n = len(y)
    dw = 1/n * np.dot(X.T,(y_pred - y))
    db = 1/n * np.sum(y_pred - y)

    return dw, db
```

```python
learning_rate = 0.002
epochs = 3000
w = np.zeros((X.shape[1], 1))
b = 0

for i in range(epochs):
    y_pred = hypothesis(w, X, b)
    
    model_loss = loss(y_pred, y, X)

    dw, db = gradient(y_pred, y, X)

    w = w - learning_rate * dw
    b = b - learning_rate * db

    if (i % 100 == 0):
        print(f"Iteration: {i}, Loss: {model_loss}")
```

```python
line = hypothesis(w, X, b)

x_vals_idx = X[:, 0].argsort()
line_sorted = line[x_vals_idx]
x_vals = X[x_vals_idx]

plt.scatter(X[:, 0], y, label="Actual Data")
plt.plot(x_vals[:, 0], line_sorted[:, 0], label="Model Output (raw)", color='red')

plt.xlabel("Feature")
plt.ylabel("Probability")
plt.legend()
plt.show()
```

```python
x_vals_idx = X[:, 1].argsort()
line_sorted = line[x_vals_idx]
x_vals = X[x_vals_idx]

plt.scatter(X[:, 1], y, label="Actual Data")
plt.plot(x_vals[:, 1], line_sorted[:, 0], label="Model Output (raw)", color='red')

plt.xlabel("Feature")
plt.ylabel("Probability")
plt.legend()
plt.show()
```
